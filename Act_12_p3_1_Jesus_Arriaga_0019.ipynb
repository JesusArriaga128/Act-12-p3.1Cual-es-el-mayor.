## Las variables
1. A --> Primer valor proporcionado
2. B --> Segundo valor proporcionado
3. M --> Variable que almacena el valor Mayor
## Formula
$Mayor = A > B$
## Diagrama de flujo LicidChart
<img src="https://raw.githubusercontent.com/JesusArriaga128/Act-12-p3.1Cual-es-el-mayor./refs/heads/main/DF%20Mayor%20de%20dos%20numero.png" width="350" height="500">


```python
# Para cuando es verdadero (Si)
# Determinando el mayor de dos números usando if else
# Datos de entrada A , B
# Mostrar el mayor M
# Comparación lógica: Si A es mayor que B

A = float(input("Ingrese el valor de A: "))
B = float(input("Ingrese el valor de B: "))

if A > B:
    M = A
    print("El mayor es A:", M)
else:
    M = B
    print("El mayor es B:", M)

print("fin de programa")
print("Programa realizado Jesus Arriaga NL = 5")
```
